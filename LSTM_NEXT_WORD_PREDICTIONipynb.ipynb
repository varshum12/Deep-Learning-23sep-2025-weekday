### convert target variable  in one  hot encoding

````python
from keras.utils  import to_categorical
y1  =  to_categorical(y)
````

````python
y1
````

#### Model  Building

````python
from keras.models import Sequential
from keras.layers  import Dense  ,  LSTM  , Embedding, Dropout
````

````python
nn  =  Sequential()
````

````python
### add  layers
nn.add(Embedding(voc_size , 10 , input_length = 3 ))

nn.add(LSTM(units = 1000 , return_sequences= True ))

nn.add(LSTM(1000 , return_sequences=True , ))

nn.add(LSTM(1000))

nn.add(Dense(500 , activation= 'relu'))

nn.add(Dense(500 , activation='relu'))



## add output layer
nn.add(Dense(voc_size , activation='softmax'))
````

````python
### model compilation
````

````python
nn.compile(optimizer  =  'adam' ,  loss  =  'categorical_crossentropy' ,  metrics  =  ['accuracy'])
````

### model  fiting

````python
hist  =  nn.fit(x1  , y1 ,  epochs  =  100)
````

````python
import matplotlib.pyplot as plt
````

````python
plt.plot(hist.history['loss'])
````

````python
def predict_word(tok,model,text):
    word=tok.texts_to_sequences([text])
    word=np.array(word)

    pred=np.argmax(model.predict(word))
    predicted_word=' '

    for key,value in tok.word_index.items():
        if value==pred:
            predicted_word=key

    print('Predictd_word:',predicted_word)
    return predicted_word
````
